```python
%matplotlib notebook
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_table('data.txt')
```

```python
data.head()
```

```python
data.shape
```

```python
data_unique = dict(zip(data.sepsis_label.unique(),data.sepsis_name.unique()))
print(data_unique)
```

```python
X = data[['glucose', 'Hemoglobin','capillary_ph','color_score']]
y = data['sepsis_label']

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
```

```python
print(X_train)
print("\n")
print(y_train)
print("\n")
print(X_test)
print("\n")
print(y_test)
```

```python
from matplotlib import cm
cmap = cm.get_cmap('gnuplot')
scatter = pd.scatter_matrix(X_train , c=y_train , marker = 'o',s=40, hist_kwds= {'bins':15} , figsize=(12,12), cmap=cmap)
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = fig.add_subplot(111,projection = '3d')
ax.scatter(X_train['glucose'], X_train['Hemoglobin'],X_train['color_score'], c=y_train ,marker = 'o',s=100)
ax.set_xlabel('glucose')
ax.set_ylabel('Hemoglobin')
ax.set_zlabel('color_score')
plt.show()
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors = 1)
```

```python
knn.fit(X_train, y_train)
```

```python
knn.score(X_test,y_test)
```

```python
data_prediction = knn.predict([[155,6.43,9.5,1]])
data_unique[data_prediction[0]]
```
